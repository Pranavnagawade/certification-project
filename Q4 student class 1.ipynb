```python

# Assuming that the column representing attendance is named "attendance"
low_attendance = stud_df[stud_df['Attended Count'] <= 4]


low_attendance.plot(kind='bar', x='Student Roll Num', y='Attended Count', color='red')

plt.title("Students with low attendance")
plt.xlabel("Student Roll Number")
plt.ylabel("Attendance less than 4")

plt.show()



#more attension
```

```python
low_attendance = stud_df[stud_df['rating_count_avg'] <= 70]
low_attendance.plot(kind='bar', x='Student Roll Num', y='rating_count_avg', color='red')
plt.title("Students are Unsatisfied")
plt.xlabel("Student Roll Number")
plt.ylabel("Rating count avg")
plt.show()
```

```python

low_attendance = stud_df[stud_df['rating_count_avg'] > 80]
low_attendance.plot(kind='bar', x='Student Roll Num', y='rating_count_avg', color='red')
plt.title("Students are Very Satisfied")
plt.xlabel("Student Roll Number")
plt.ylabel("Rating count avg")

plt.show()
```

```python
# averge Student ratng 
low_attendance = stud_df[(stud_df['rating_count_avg'] >= 75) & (stud_df['rating_count_avg']<=80)]
low_attendance.plot(kind='bar', x='Student Roll Num', y='rating_count_avg', color='blue')
plt.title("Students are average performace ")
plt.xlabel("Student Roll Number")
plt.ylabel("Rating count avg")

#axs[1]plt.show()
```
